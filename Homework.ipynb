```python
import matplotlib.pyplot as plt
students_copy.dropna(inplace=True)
display(students_copy['gender'].value_counts())
fig = plt.figure(figsize=(7,3))
ax = fig.add_axes([0, 0, 0.5, 1])
bar_categories = ['Male','Female']
bar_values = (students_copy['gender'].value_counts().values)
ax.bar(bar_categories,bar_values,color='c')
ax.set_title("Male and Female students distribution")
fig.tight_layout()
plt.show()
```
